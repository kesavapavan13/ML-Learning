```python
# Library used for this file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import mutual_info_classif,chi2
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSet/banana_quality_dataset.csv")
```

```python
df.isnull().sum()
```

```python
# divides x and y here itself
x=df.drop(columns=["sample_id","quality_category","harvest_date"],axis=1)
y=df["quality_category"]
```

```python
x.shape
```

```python
x.head(1)
```

# Columns Transformation for Encoding
* before Fetaure selection we have complete the Endcoding

```python
transformer = ColumnTransformer(transformers=[("Cat",OrdinalEncoder(),[0,1,4,]),
                                              ("numeric",MinMaxScaler(),[2,3,5,6,7,8,9,10,11,12])])
x_transformed = transformer.fit_transform(x)
```

```python
x_transformed = pd.DataFrame(x_transformed,columns=x.columns)
x_transformed.columns
```

# Mutual Information
* Mutual Information measures how much information a feature provides about the target.

```python
mutual_score = mutual_info_classif(x_transformed,y)
mutual_info = pd.DataFrame({"Feature":x_transformed.columns,
                            "Mutual_Score": mutual_score}).sort_values(by="Mutual_Score",ascending=False)
mutual_info
```

# Chi-Square
* The Chi-Square test measures whether a categorical feature is statistically dependent on the target variable.

```python
chi_score = chi2(x_transformed,y)

chi_info = pd.DataFrame({"Feature":x_transformed.columns,
                            "Chi_Score": chi_score[0]}).sort_values(by="Chi_Score",ascending=False)
chi_info
```
