```python
sent = "Customers find this refrigerator to be of good quality and value for money with positive feedback about its design storage capacity and cooling performance However the functionality receives mixed reviews with some reporting excellent performance while others say it suddenly stops working ðŸ˜¡"
```

## Lower

```python
sent = sent.lower()
sent
```

## Removeing puncation and special charaters

```python
import re
sent = re.sub(r'[^a-z\s]','',sent)
print(sent)
```

## Tokenization

```python
tokens = sent.split()
print(tokens)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# by using tokenization
from nltk.tokenize import word_tokenize
tokens = word_tokenize(sent)
print(tokens)
```

## Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
len(stop_words) #number of the stopwords
```

```python
tokens = [word for word in tokens if word not in stop_words]
print(tokens)
```

## PorterStemmer

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmed_tokens = [stemmer.stem(word) for word in tokens]
print(stemmed_tokens)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
print(lemmatized_tokens)
```
