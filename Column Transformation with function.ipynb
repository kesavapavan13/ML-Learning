```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSet/loan_approved1.csv")
df.head()
```

```python
#check the null values in DataSet
df.isnull().sum()
```

```python
# supperate the target variable
y = df['Loan_Status (Approved)']
x = df.drop('Loan_Status (Approved)',axis=1)
```

```python
# divide into numerical and catregorical columns
numeric_col = x.select_dtypes(include=np.number).columns
cat_col = x.select_dtypes(exclude=np.number).columns
```

```python
for i in cat_col:
  print(x[i].unique())
```

```python
def column_Trans(df):
  # divide into numerical and catregorical columns
  numeric_col = df.select_dtypes(include=np.number).columns
  cat_col = df.select_dtypes(exclude=np.number).columns

  return ColumnTransformer(transformers = [("num",StandardScaler(),numeric_col), #Feature Scaling
                                           ("Ord",OrdinalEncoder(categories=[[ 'Rural','Semiurban','Urban']]),['Property_Area']),# To keep in order
                                          ("cat",OneHotEncoder(sparse_output=False),cat_col)], # Encoding
                                          remainder="passthrough")
```

```python
per =column_Trans(x)
```

```python
prep = per.fit_transform(x)
```

```python
df_1=pd.DataFrame(prep)
```
