```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import re
def Get_words(df):
  df['text'] = df['text'].str.lower()
  df['text'] = df['text'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
  return df
```

```python
df=Get_words(df)
```

```python
df['Tokens'] = df['text'].str.split()
```

```python
df['Len_token_stopwords'] = df['Tokens'].apply(lambda x: len(x))
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
df['token_not_stopwords'] = df['Tokens'].apply(lambda x: [word for word in x if word not in stopwords.words('english')])
```

```python
df['Len_token_not_stopwords'] = df['token_not_stopwords'].apply(lambda x: len(x))
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
df['stemming'] = df['token_not_stopwords'].apply(lambda x: [ps.stem(word) for word in x])
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
df['lemmatization'] = df['token_not_stopwords'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
```

```python
df.head()
```

```python
df['sent_lemmatization'] = df['lemmatization'].apply(lambda x: ' '.join(x))
```

```python
df.head()
```

# Bag Of Words(BOW)
## Binary Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary=True)
bag = vectorizer.fit_transform(df['sent_lemmatization'])
```

```python
print(vectorizer.get_feature_names_out())
print(bag.toarray())
```

```python
vectorizer1 = CountVectorizer()
bag1 = vectorizer1.fit_transform(df['sent_lemmatization'])
```

```python
print(vectorizer1.get_feature_names_out())
print(bag1.toarray())
```
