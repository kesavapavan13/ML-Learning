```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = ["I love ML and ML is part of Data Science",
          "I love DL and Dl is sub-part of ML and Data science ",
          "I love NLP and it used for Text","NLP,ML,DL are used in Data Science",
          "I Love to learn Data Science and data science persent of ML,DL,NPL"]
```

# Bag Of Words(BOW)
## Binary Bag of words

```python
Vectorizer = CountVectorizer(binary=True)
bag = Vectorizer.fit_transform(corpus)
```

```python
print(Vectorizer.get_feature_names_out())
print(bag.toarray())
```

## Count Bag of Words

```python
Vectorizer1 = CountVectorizer()
bag1 = Vectorizer1.fit_transform(corpus)
```

```python
print(Vectorizer1.get_feature_names_out())
print(bag1.toarray())
```

# TF-IDF

```python
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(corpus)
```

```python
print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", x.toarray())
```
